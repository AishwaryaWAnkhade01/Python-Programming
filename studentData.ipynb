 #### What is the average family income of the student?

#### Aim: Convert income brackets to numbers and find average.

#### Approach: Map categories to mid-point numerical values.

```python

income_mapping = {
    "0-2 Lakh": 1,
    "2-5 Lakh": 3.5,
    "5-7 Lakh": 6,
    "7 Lakh+": 7
}

# Convert "Family Income" column using mapping
cl_df['Family Income Numeric'] = cl_df['Family Income'].map(income_mapping)

# Calculate the average family income
average_income = cl_df['Family Income Numeric'].mean()

print(f"Average Family Income: {average_income:.2f} Lakh")
```

##### Insight: Average family income is approx. {avg_income:.2f} LPA.

##### Conclusion: Useful for pricing workshops or financial support programs.

#### How does the expected salary vary based on CGPA, Family income, Python experience?

#### Aim: To analyze how expected salary is influenced by GPA, income, and experience.

```python
income_map = {
    "0-2 Lakh": 1,
    "2-5 Lakh": 3.5,
    "5-7 Lakh": 6,
    "7 Lakh+": 8
}
cl_df['Family Income Numeric'] = df['Family Income'].map(income_map)

corr_data = cl_df[['Expected salary (Lac)', 'CGPA', 'Experience with python (Months)', 'Family Income Numeric']]

# Correlation Heatmap
sns.heatmap(corr_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

##### Insight: Students with higher GPA and more python experience expect higher salaries.

##### Conclusion: Skill-building and academic performance correlate with salary expectations.

#### Which event tends to attract more students from specific fields of study?

#### Aim: Identify which events are popular among different designations or colleges.

```python
pivot=cl_df.pivot_table(index='Events',columns='Designation',aggfunc='size',fill_value=0)
plt.figure(figsize=(14,8))
sns.heatmap(pivot,cmap="YlGnBu",annot=True,fmt='d')
plt.title("Event Participation by designation ")
plt.show()
```

##### Insight: Data Science and tech-related events attract the most Engineering and IT students.

##### Conclusion: Designation-specific marketing can boost relevant event turnout.

####  Do students in leadership positions have higher GPAs or salary expectations?

#### Aim: Assess the academic and financial impact of leadership roles.

```python
leadership_stats=df.groupby('Leadership- skills')[['CGPA','Expected salary (Lac)']].mean()
leadership_stats.plot(kind='bar',figsize=(98,25),colormap='Set2')
plt.title("Leadership Skills vs Avg CGPA and Expected Salary")
plt.ylabel("Average")
plt.xticks(rotation=0)
plt.show()
```

##### Insight: Students in leadership roles tend to have marginally higher CGPAs and salary expectations.

##### Conclusion: Leadership development might contribute positively to career goals.

#### How many students are graduating by the end of 2024?

#### Aim: Find number of students eligible to graduate in 2024 or earlier.

```python
# Add new column for graduation status
cl_df['Graduating by 2024'] = cl_df['Year of Graduation'] <= 2024

# Count how many True (Yes) and False (No)
counts = cl_df['Graduating by 2024'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
counts.plot.pie(
    autopct='%1.1f%%',
    labels=["Yes", "No"] if counts.index[0] else ["No", "Yes"],  # Just in case order is flipped
    colors=['lightgreen', 'lightcoral']
)
plt.title("Students Graduating by End of 2024")
plt.ylabel("")
plt.show()
```

##### Insight: A large portion of students are about to graduate soon.

##### Conclusion: Focus should be placed on this cohort for placements and career counseling.

#### Students with high CGPA & Python experience who had high salary expectations (Avg)

#### Aim: Analyze average salary expectations for top-performing students.

```python
top_perf = df[
    (df['CGPA'] > 8) &
    (df['Experience with python (Months)'] > 6) &
    (df['Expected salary (Lac)'] > 6)
]

avg_stats = top_perf[['CGPA', 'Experience with python (Months)', 'Expected salary (Lac)']].mean()
avg_stats.plot(kind='bar', color='seagreen')
plt.title("Average Stats for Top Performers")
plt.ylabel("Average")
plt.xticks(rotation=45)
plt.show()
```

##### Insight: High performers expect ~{avg_salary:.2f} LPA on average.

##### Conclusion: Companies targeting skilled freshers can align offerings accordingly.
